### Importing `customer_basket`  and `final_clusters` files 

```python
# load the cluster DataFrame from a pickle file
cl_f1 = pd.read_csv('files_toimport/final_clusters.csv')
cust_basket = pd.read_csv('customer_basket.csv')
```

```python
# create a clusters basket dataframe which contains the 'customer_id', 'list_of_goods' and ' final_cl'
cl = cust_basket[['customer_id']]
cl = cl.merge(cl_f1, on='customer_id', how='left')
cl = cl.merge(cust_basket, on='customer_id')
cl.drop('invoice_id', axis=1, inplace=True)
```

```python
# dividing clusters basket dataframe into individual cluster dataframes with the same columns
cl_4 = cl[cl['final_cluster'] == 'Fake Vegetarians']
cl_3 = cl[cl['final_cluster'] == 'Millenials']
cl_2 = cl[cl['final_cluster'] == 'Gamer Parents']
cl_1 = cl[cl['final_cluster'] == 'Misers']
cl0 = cl[cl['final_cluster'] == 'Local Folks']
cl1 = cl[cl['final_cluster'] == 'Youngsters']
cl2 = cl[cl['final_cluster'] == 'Balanced Parents']
cl3 = cl[cl['final_cluster'] == 'Karens'] # no invoices
cl4 = cl[cl['final_cluster'] == 'Petters'] # no invoices
cl5 = cl[cl['final_cluster'] == 'Settled Techies']
cl6 = cl[cl['final_cluster'] == 'Dirty Antiques'] # no invoices
cl7 = cl[cl['final_cluster'] == 'Stingy Pescatarians']
cl8 = cl[cl['final_cluster'] == 'Retired Elders'] 
```

```python
# checking whether all the above dataframes have invoices/transactions
#cl8
```

```python
# creating a list containing only the3 'list_of_goods' column of each cluster
items_4 = cl_4['list_of_goods'].to_list()
items_4 = [ast.literal_eval(_) for _ in items_4]

items_3 = cl_3['list_of_goods'].to_list()
items_3 = [ast.literal_eval(_) for _ in items_3]

items_2 = cl_2['list_of_goods'].to_list()
items_2 = [ast.literal_eval(_) for _ in items_2]

items_1 = cl_1['list_of_goods'].to_list()
items_1 = [ast.literal_eval(_) for _ in items_1]

items0 = cl0['list_of_goods'].to_list()
items0 = [ast.literal_eval(_) for _ in items0]

items1 = cl1['list_of_goods'].to_list()
items1 = [ast.literal_eval(_) for _ in items1]

items2 = cl2['list_of_goods'].to_list()
items2 = [ast.literal_eval(_) for _ in items2]

items4 = cl4['list_of_goods'].to_list()
items4 = [ast.literal_eval(_) for _ in items4]

items5 = cl5['list_of_goods'].to_list()
items5 = [ast.literal_eval(_) for _ in items5]

items7 = cl7['list_of_goods'].to_list()
items7 = [ast.literal_eval(_) for _ in items7]

items8 = cl8['list_of_goods'].to_list()
items8 = [ast.literal_eval(_) for _ in items8]
```

### Apriori Algorithm for each cluster

```python
te = TransactionEncoder()
```

#### Cluster -4: `Fake Vegetarians`

```python
# splitting the 'list_of_goods' for the -4 cluster
i_4_tr = items_4[:int(len(items_4)*0.8)]
i_4_t = items_4[int(len(items_4)*0.8):]

# fitting the train set using TransactionEncoder
te_fit_4 = te.fit(i_4_tr).transform(i_4_tr)
i_4 = pd.DataFrame(te_fit_4, columns=te.columns_)
```

```python
i_4.head(10)
```

```python
# checking the itemsets with a support >= 0.2
i_4_frq = apriori(i_4, min_support=0.2, use_colnames=True)
i_4_frq.sort_values(by='support', ascending=False)
```

```python
# checking the rules based on the confidence >= 0.6
i_4_rules = association_rules(i_4_frq, metric="confidence", min_threshold=0.6)
i_4_rules
```

```python
# decrease support and confidence to get rows with higher lift
i_4_frq_ = apriori(i_4, min_support=0.05, use_colnames=True)
i_4_rules_ = association_rules(i_4_frq_, metric="confidence", min_threshold=0.5)
i_4_rules_.sort_values(by='lift', ascending=False).head(10)
```

#### Promotions for `Fake Vegetarians`:
- _Soup Lovers_: 20% off on the following: carrots, tomatoes, frozen veggies and asparagus!
- _Tomto Peers_: Get 30% off on tomatoes!
